```python
baloes = Image.open('baloes.jpg')
baloes
```

```python
def escala_cinza(imagem_colorida):
    w, h = imagem_colorida.size
    img = Image.new('RGB', (w, h))
    for x in range (w):
        for y in range(h):
          pxl = imagem_colorida.getpixel((x, y))
          # RGB multiplicado pelos pesos de iliminação para cada cor respectivamente
          lum = int(0.3 * pxl[0] + 0.59 * pxl[1] + 0.11 * pxl[2])  
          img.putpixel((x, y), (lum, lum, lum))

    return img



def image_pb(escala_cinza): 
    threshold = 128  
    img_pb = baloes_gs.point(lambda p: p > threshold and 255)
    
    return img_pb



def img_resize(img):
    img = baloes.resize((500, 286))
    
    return img
```
